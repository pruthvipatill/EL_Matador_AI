```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\PRUTHVI\OneDrive\Desktop\EL Matador AI\data\ufc-master.csv")
```

```python
df = df.drop(columns=['location','empty_arena'])
```

```python
df.select_dtypes(include=['object'])
```

```python
#categorizing the data into numerical columns and text columns
#filling the nan numerical columns with 0 and text with unknown
nums_col = df.select_dtypes(include =['int64','float64']).columns
df[nums_col]=df[nums_col].fillna(0)
```

```python
text_col = df.select_dtypes(include=['object']).columns
df[text_col] = df[text_col].fillna('unknown')
```

```python
for row in df['date']:
    print(row)
```

```python
df['red_win'] = (df['Winner']=='Red').astype(int)
```

```python
df = df.drop(columns=['Winner'])
```

```python
df = df.copy()
```

```python
df = pd.get_dummies(df,columns=['R_Stance','B_Stance'],drop_first=True)
```
